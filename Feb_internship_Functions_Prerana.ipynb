```python
def parking_system(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1

        if current > peak:
            peak = current

    status = "Available"
    if current > capacity:
        status = "Full"

    print("Currently Parked Vehicles:", current)
    print("Peak Usage:", peak)
    print("Parking Status:", status)


# Sample
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_system(capacity, logs)
```

```python
def delivery_time(distance, traffic, weather):
    speed = 30  # km per hour
    time_minutes = (distance / speed) * 60

    # Traffic delay
    if traffic == "High":
        time_minutes += 15
    elif traffic == "Medium":
        time_minutes += 10
    else:
        time_minutes += 5

    # Weather delay
    if weather == "Rainy":
        time_minutes += 10
    elif weather == "Storm":
        time_minutes += 20

    print("Estimated Delivery Time:", int(time_minutes), "minutes")


# Sample
delivery_time(8, "High", "Rainy")
```

```python
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)


# Sample
total_seats = 200
booked_seats = [1] * 150
seat_occupancy(total_seats, booked_seats)
```

```python
def server_status(cpu_readings):
    total = 0
    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)


# Sample
readings = [45, 60, 70, 85, 90]
server_status(readings)
```

```python
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
resource_monitor(usage)
```

```python
def event_registration(capacity, registrations):
    confirmed = 0
    waitlist = 0

    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1

    # Registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
capacity = 100
registrations = 105

event_registration(capacity, registrations)
```
